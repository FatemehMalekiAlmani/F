```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# GeneticAlgorithm solver for linear program
#    max   3x + 5xy
#    s.t.  x <= 4
#          2*y <= 12
#          3 *x+ 2*y <= 18
#          0<= x <= 10
#          0<= y <= 10
#          x is integer
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
!pip install geneticalgorithm
import numpy as np
from geneticalgorithm import geneticalgorithm as ga

# _______________________________Function_______________________________
def f(x):
    pen=0
    if not x[0] <= 4 : pen=np.inf
    if not 2*x[1] <= 12 : pen=np.inf
    if not 3*x[0]+ 5* x[1] * x[0] <= 18 :pen=np.inf
    return -(3*x[0]+5*x[1]*x[0])+ pen
#_______________________________variables bounds _______________________________
varbounds= np.array([[0,10], [0,10]])
#_______________________________kind of variables_______________________________
vartype=np.array([['int'], ['real']])
#_______________________________model_______________________________
model= ga(function= f, dimension=2, variable_type_mixed= vartype, variable_boundaries= varbounds)
#_______________________________results_______________________________
model.run()
```
